```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("HIV DATASET 2.csv")
df
```

```python
df.drop([0,1,2,3], axis=0,inplace = True)
cols = ["PUBCHEM_EXT_DATASOURCE_SMILES", "PUBCHEM_ACTIVITY_OUTCOME"]
df = df[cols].reset_index()
df.drop("index", axis=1, inplace = True)
df.columns = ["SMILES", "ACTIVITY_OUTCOME"]
df.dropna(axis=0)
df
```

```python
def binarize(x):
    if x == "Active":
        return 1
    else:
        return 0
```

```python
df["ACTIVITY"] = df["ACTIVITY_OUTCOME"].apply(binarize)
df.drop("ACTIVITY_OUTCOME", axis=1, inplace = True)
df
```

```python
df_new = df.copy()
#the data entries duplicates are dropped
df_new.drop_duplicates(inplace = True, ignore_index = True)
df_new.dropna(axis=0, inplace = True)
df_new
```

```python
df_new.to_csv("HIV bigdata.csv", index=False)
```
